```python
def get_state_and_year(state, time):
    dataframe = df.loc[(df['country'] == state) & (df['year'] == time)]
    return dataframe.sort_values(by='age')
```

```python
states = df['country'].unique()
def plot_states_by_year(year):
    tot_pos = []
    valid_states = []
    
    for i in states:
        if get_state_and_year(i, year).shape[0] != 0:
            valid_states.append(i)
            
    semi_states = math.ceil(len(valid_states)/2)
            
    for n in range(len(valid_states)):
        if n < semi_states:
            tot_pos.append((n, 0) if n%2 == 0 else (n, 1))
        else:
            tot_pos.append((n-semi_states, 0) if n%2 == 0 else (n-semi_states, 1))
    
    tot_pos = sorted(tot_pos)
    f, axis = plt.subplots(semi_states, 2, figsize = (28, 3*semi_states))
    for n, state in enumerate(valid_states):
        sns.barplot(data = get_state_and_year(state, year), x='age', y='suicides/100k pop', hue='sex', ax=axis[tot_pos[n]])
        axis[tot_pos[n]].set_title(state)

```

```python
def plot_years_by_state(state):
    tot_pos = []
    valid_years = []
    
    for i in years:
        if get_state_and_year(state, i).shape[0] != 0:
            valid_years.append(i)
            
    semi_states = math.ceil(len(valid_years)/2)
            
    for n in range(len(valid_years)):
        if n < semi_states:
            tot_pos.append((n, 0) if n%2 == 0 else (n, 1))
        else:
            tot_pos.append((n-semi_states, 0) if n%2 == 0 else (n-semi_states, 1))
    
    tot_pos = sorted(tot_pos)
    f, axis = plt.subplots(semi_states, 2, figsize = (28, 3*semi_states))
    for n, year in enumerate(valid_years):
        sns.barplot(data = get_state_and_year(state, year), x='age', y='suicides/100k pop', hue='sex', hue_order=['female', 'male'], ax=axis[tot_pos[n]])
        axis[tot_pos[n]].set_title(str(state) + ' in ' + str(year))

```
